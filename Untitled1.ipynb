```python
users_df = pd.read_csv('combined_data(2025-3-1-2026-2-28) by channel and user.csv')
def get_all_channels(metric):

    df = users_df.copy()

    grouped = df.groupby("Channel").agg(
        uploaded=("Uploaded Count","sum"),
        published=("Published Count","sum")
    ).reset_index()

    grouped["value"] = grouped[metric]
    grouped["name"] = grouped["Channel"]

    return grouped.sort_values("value", ascending=False) \
        .head(20).to_dict("records")

df = get_all_channels("uploaded")
print(df)
```

```python
client_df = pd.read_csv('CLIENT 1 combined_data(2025-3-1-2026-2-28).csv')

def get_channels_by_client(client, metric="published"):

    df = client_df.copy()

    # normalize
    if "Channel" not in df.columns:
        return []

    df["Channel"] = df["Channel"].astype(str).str.strip()

    # filter by client (if multiple clients later)
    if "Client" in df.columns:
        df = df[df["Client"] == client]

    df["uploaded"] = df["Uploaded Count"]
    df["published"] = df["Published Count"]

    df["value"] = df[metric]
    df["name"] = df["Channel"]

    return df[["name","uploaded","published","value"]] \
        .sort_values("value", ascending=False) \
        .head(20) \
        .to_dict("records")

df2 = get_channels_by_client('Client1', "uploaded")

print(df2)
```
